```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import string

# NLP libraries
import nltk
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer

# Word Cloud
from wordcloud import WordCloud

# Download required NLTK data (Run only once)
nltk.download('stopwords')
nltk.download('vader_lexicon')
```

```python
# Load the dataset
df = pd.read_csv(r"C:\Users\Admin\Desktop\NLP\YoutubeCommentsDataSet.csv")

# Display first 5 rows
print(df.head())

# Display column names
print(df.columns)
```

```python
# Keep only the comment column
df = df[['Comment']]

print(df.head())
```

```python
print(df.columns)
```

```python
# Function to clean text
def clean_text(text):

    # Convert to lowercase
    text = str(text).lower()

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Remove stop words
    stop_words = set(stopwords.words('english'))

    words = text.split()

    cleaned_words = []

    for word in words:
        if word not in stop_words:
            cleaned_words.append(word)

    return " ".join(cleaned_words)

# Apply cleaning
df["Clean_Comment"] = df["Comment"].apply(clean_text)

print(df.head())
```

```python
# Create sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Function to find sentiment
def analyze_sentiment(text):

    score = sia.polarity_scores(text)

    if score['compound'] >= 0.05:
        return "Positive"

    elif score['compound'] <= -0.05:
        return "Negative"

    else:
        return "Neutral"

# Apply sentiment analysis
df["Sentiment"] = df["Clean_Comment"].apply(analyze_sentiment)

print(df.head())
```

```python
# Count sentiments
sentiment_count = df["Sentiment"].value_counts()

print(sentiment_count)
```

```python
# Plot bar chart
sentiment_count.plot(kind="bar")

plt.title("Public Opinion on YouTube Comments")

plt.xlabel("Sentiment")

plt.ylabel("Number of Comments")

plt.show()
```

```python
# Pie chart
sentiment_count.plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Sentiment Distribution")

plt.ylabel("")

plt.show()
```

```python
# Join all cleaned comments
text = " ".join(df["Clean_Comment"])

# Generate word cloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white"
).generate(text)

# Display word cloud
plt.figure(figsize=(10,5))

plt.imshow(wordcloud)

plt.axis("off")

plt.title("Trending Topics")

plt.show()
```

```python
# Find most frequent words
words = text.split()

frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Sort words by frequency
sorted_words = sorted(frequency.items(), key=lambda x: x[1], reverse=True)

print("Top 10 Trending Words")

for word, count in sorted_words[:10]:
    print(word, ":", count)
```
